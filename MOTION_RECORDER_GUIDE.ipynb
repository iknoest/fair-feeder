## Section 3: Object Detection with ai-edge-litert (NOT tflite-runtime)

### The Difference: TFLite Runtimes
| Feature | tflite-runtime | ai-edge-litert |
|---------|---|---|
| **Creator** | Google (old) | Google (NEW 2024) |
| **Speed** | Slower | ✅ **Faster** |
| **Raspberry Pi** | ❌ Hard to install | ✅ **Easy to install** |
| **API** | `.set_tensor()` + `.invoke()` | `.invoke(input)` |

**We switched to ai-edge-litert because:**
- ✅ Easier to install on Raspberry Pi
- ✅ Better performance for edge devices
- ✅ Official Google recommendation for 2024

### Why We Need Object Detection
- **Motion detection** finds ANY movement (birds, wind, bugs)
- **Object detection** identifies WHAT moved (is it a cat?)
- **Combination** = cat videos only! 🐱
